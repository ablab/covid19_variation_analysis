```python
### Compute histogram bins

bins = 30000/300
bin2freq_s = Counter()
bin2freq_t = Counter()
bin2freq_m = Counter()

c = Counter()
N = 0
for x, freq in multiples.most_common(100000):
    c[freq] += 1
    N += 1

for x, freq in multiples.most_common(100000):
    if freq == 1:
       bin2freq_s[x / 300]  += 1
    if 1 < freq:
       bin2freq_t[x / 300]  += 1
    if freq > 10:
       bin2freq_m[x / 300]  += 1
```

```python
### draw image 

scale = 35
shift = 50

canvas = Canvas(width=1000, height=900)
canvas.fill_style = white
canvas.fill_rect(0, 0, 1000, 900)
canvas.fill_style = black
    
def draw_utr(canvas, start, length, color, shift_to_right=50):
    
    canvas.fill_style = color
    canvas.begin_path()
    canvas.move_to(start+shift_to_right, 90)
    canvas.line_to(start+length+shift_to_right, 90)
    canvas.line_to(start+length+shift_to_right, 60)
    canvas.line_to(start+shift_to_right, 60)
    canvas.line_to(start+shift_to_right, 90)
    canvas.fill()

def draw_gene(canvas, start, length, color, shift_to_right=50):
    
    canvas.fill_style = color
    canvas.begin_path()
    canvas.move_to(start+shift_to_right, 100)
    canvas.line_to(start+length+shift_to_right, 100)
    canvas.line_to(start+length+shift_to_right, 50)
    canvas.line_to(start+shift_to_right, 50)
    canvas.line_to(start+shift_to_right, 100)
    canvas.fill()
    
def draw_nsp(canvas, start, length, color, shift_to_right=50):
    
    canvas.fill_style = color
    canvas.begin_path()
    canvas.move_to(start+shift_to_right, 100-5)
    canvas.line_to(start+length+shift_to_right, 100-5)
    canvas.line_to(start+length+shift_to_right, 50+5)
    canvas.line_to(start+shift_to_right, 50+5)
    canvas.line_to(start+shift_to_right, 100-5)
    canvas.fill()
    
def draw_brackets(canvas, veritcal_pos):
    
    line_width = 3
    
    h_start = 110
    canvas.fill_style = black
    H = 200
    space = 20
            
    canvas.fill_rect(25, h_start, 20, line_width)
    canvas.fill_rect(25, h_start, line_width, H/2-5)
    canvas.fill_rect(25, h_start+H/2-5, 20, line_width)
    
    canvas.fill_rect(25, h_start+H/2+5, 20, line_width)
    canvas.fill_rect(25, h_start+H/2+5, line_width, H/2-5)
    canvas.fill_rect(25, h_start+H/2+5+H/2-5, 20, line_width)
    canvas.fill_rect(25, h_start+H/2+5+(H/2-5)/2, 10, line_width)

    canvas.fill_rect(25, h_start+H+space, 20, line_width)
    canvas.fill_rect(25, h_start+H+space, line_width, H/2-5)
    canvas.fill_rect(25, h_start+H/2+H+space-5, 20, line_width)
    
    canvas.fill_rect(25, h_start+H+space+H/2+5, 20, line_width)
    canvas.fill_rect(25, h_start+H+space+H/2+5, line_width, H/2-5)
    canvas.fill_rect(25, h_start+H/2+H+space+H/2, 20, line_width)
    canvas.fill_rect(25, h_start+H+space+H/2+5 + H/4 , 10, line_width)
    
    canvas.fill_rect(25, h_start+2*(H+space), 20, line_width)
    canvas.fill_rect(25, h_start+2*(H+space), line_width, H/2-5)
    canvas.fill_rect(25, h_start+H/2+2*(H+space)-5, 20, line_width)
    
    canvas.fill_rect(25, h_start+2*(H+space)+H/2+5, 20, line_width)
    canvas.fill_rect(25, h_start+2*(H+space)+H/2+5, line_width, H/2-5)
    canvas.fill_rect(25, h_start+H/2+2*(H+space)+H/2, 20, line_width)
    canvas.fill_rect(25, h_start+2*(H+space)+H/2+5 + H/4, 10, line_width)
  

with hold_canvas(canvas):

    prev_nsp = None
    for start, end, gene, color in genes:
        gene = gene[0]
        length = (end-start)
        start = start/scale
        length = length/scale
        if "UTR" in gene:
            draw_utr(canvas, start, length, color)
        elif "nsp" in gene:        
            if prev_nsp:
                start = prev_nsp + 1
                if prev_nsp+length == 615: ### special case for last subprotein
                    length -= 5
            draw_nsp(canvas, start, length, color)
            prev_nsp = start+length
        else:
            draw_gene(canvas, start, length, color)

    draw_brackets(canvas, 0)

    n_singl = 0
    n_mult = 0
    n_ten = 0

    for x, freq in multiples.most_common(100000):
        if freq == 1:
            _x = x/scale + shift
            canvas.fill_rect(_x, 110, 2, 100-3)
            n_singl += 1
        if freq > 1:
            _x = x/scale + shift
            canvas.fill_rect(_x, 110+1*(200+20), 2, 100-3)
            n_mult += 1
        if freq > 10:
            _x = x/scale + shift
            canvas.fill_rect(_x, 110+2*(200+20), 2, 100-3)
            n_ten += 1

    for x, freq in bin2freq_s.most_common(100000):
        x = (x*300)/scale + shift 

        canvas.fill_rect(x, 215, 7, freq)

    for x, freq in bin2freq_t.most_common(100000):
        x = (x*300)/scale + shift 
        canvas.fill_rect(x, 110+1*(200+20)+105, 7, freq)

    for x, freq in bin2freq_m.most_common(100000):
        x = (x*300)/scale + shift 
        canvas.fill_rect(x, 110+2*(200+20)+105, 7, freq)

    canvas.font = "28px arial"
    canvas.fill_text("orf1ab", 330, 40)
    canvas.fill_text("S", 713, 40)
    canvas.fill_text("E", 790, 40)
    canvas.fill_text("M", 810, 40)
    canvas.fill_text("N", 863, 40)

    canvas.fill_text("Singletons (%s)" % n_singl, 360, 278)
    canvas.fill_text("More than one genome (%s)" % n_mult, 300, 502)
    canvas.fill_text("More than ten genomes (%s)" % n_ten, 300, 722)

    canvas.font = "14px arial"
    canvas.fill_text("50", 30, 278)
    canvas.fill_text("50", 30, 502)
    canvas.fill_text("50", 30, 722)

canvas
```
